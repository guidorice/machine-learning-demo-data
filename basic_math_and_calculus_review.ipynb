# summations in python ( is sigma character meaning summation)  

```python
summation = sum(2*i for i in range(1,6))
print(summation)
```

```python
x = [1, 4, 6, 2]
n = len(x)

summation = sum(10*x[i] for i in range(0,n))
print(summation)
```

```python
i,n = symbols('i n')

# iterate each element i from 1 to n,
# then multiply and sum
summation = Sum(2*i,(i,1,n))

# specify n as 5,
# iterating the numbers 1 through 5
up_to_5 = summation.subs(n, 5)
print(up_to_5.doit()) # 30
```

```python
# calculating compount interest in python
from math import exp

p = 100  # principal
r = .20  # interest rate
t = 2.0  # time (years)
n = 12   # periods (months)

a = p * (1 + (r/n))**(n * t)

print(a) # prints 148.69146179463576
```

```python
# calculating continuous interest in python

p = 100 # principal, starting amount
r = .20 # interest rate, by year
t = 2.0 # time, number of years

a = p * exp(r*t)  # base e

print(a) # prints 149.18246976412703
```

# limits

```python
x = symbols('x')
expr = 1/x
plot(expr, xlim=(-10,10), ylim=(-10,10))
```

```python
# using sympy to calculate limits
x = symbols('x')
f = 1 / x
result = limit(f, x, oo)  # Note that infinity is cleverly expressed in SymPy with oo.

print(result) # 0
```

```python
# calculate Euler's number with sympy
n = symbols('n')
f = (1 + (1/n))**n
result = limit(f, n, oo)

print(result) # E
print(result.evalf()) # 2.71828182845905
```

# derivatives

```python
# plot f(x) = x ** 2
x = symbols('x')
f = x ** 2
plot(f, xlim=(-10,10), ylim=(0,10))
```

```python
# a derivative calculator in python (using simple rise-over-run formula)
def derivative_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
    return m

def my_function(x):
    return x**2

slope_at_2 = derivative_x(my_function, 2, .00001)

print(slope_at_2) # prints 4.000010000000827
```

```python
# the derivative function will make the exponent a multiplier and then decrement the exponent by 1,
```

```python
# Calculating a derivative in SymPy

# Declare 'x' to SymPy
x = symbols('x')

# Now just use Python syntax to declare function
f = x**2

# Calculate the derivative of the function
dx_f = diff(f)
print(dx_f) # prints 2*x
```

```python
for n in range(-3, 4):
    print(n, dx_f(n))
```

# partial derivatives

```python
# each variable has a gradient (slope)

# calculating partial derivatives with sympy

# Declare x and y to SymPy
x,y = symbols('x y')

# Now just use Python syntax to declare function
f = 2*x**3 + 3*y**3

# Calculate the partial derivatives for x and y
dx_f = diff(f, x)
dy_f = diff(f, y)

print(dx_f) # prints 6*x**2
print(dy_f) # prints 9*y**2

# plot the function
plot3d(f)
```

```python
plot(dx_f)
```
